```python
def adj_list(G):
    L=G.edges()
    ajd=[]
    for i in range(G.order()):
        uadj=[]
        for j in range(len(L)):
            if(L[j][0]==i):
                if(L[j][1] not in uadj):
                    uadj.append(L[j][1])
            elif(L[j][1]==i):
                if(L[j][0] not in uadj):
                    uadj.append(L[j][0])
        ajd.append(uadj)
    return ajd

GG=digraphs.RandomDirectedGNM(5, 8)
GGG=graphs.RandomGNM(5, 8)
GG.show()
print(adj_list(GG))
GGG.show()
print(adj_list(GGG))
```
