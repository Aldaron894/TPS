```python
def adj_list(G):
    n=G.order()
    L=[[] for i in range(n)]
    if G.is_directed():
        for u, v, _ in G.edges():
            L[u].append(v)
    else:
        for u, v, _ in G.edges():
            L[u].append(v)
            L[v].append(u)
    return L

GG=digraphs.RandomDirectedGNM(5, 8)
GGG=graphs.RandomGNM(5, 8)
GG.show()
print(adj_list(GG))
GGG.show()
print(adj_list(GGG))
```
